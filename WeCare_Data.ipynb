```python
WeCareData.reset_index(inplace = True)

df = pd.melt(WeCareData, id_vars = 'Date_Submitted', value_vars=['RM50_SETEL','RM100_SETEL','RM200_SETEL','RM10_PARKSON','RM50_PARKSON','RM50_AL_IKHSAN'])
df.columns = ['ds','Voucher','y']
df.head()
```

```python
group_by_voucher = df.groupby('Voucher')
group_by_voucher.groups.keys()
```

```python
def train_and_forecast(group):
  m = Prophet()

  m.fit(group)

  future = m.make_future_dataframe(periods=365)

  forecast = m.predict(future)[['ds','yhat','yhat_lower','yhat_upper']]
  forecast['Voucher'] = group['Voucher'].iloc[0]

  return forecast[['ds','Voucher','yhat','yhat_upper','yhat_lower']]

```

```python
import time
start_time = time.time()
Voucher_list=['RM50_SETEL','RM100_SETEL','RM200_SETEL','RM10_PARKSON','RM50_PARKSON','RM50_AL_IKHSAN']
for_loop_forecast = pd.DataFrame()

for index in Voucher_list:
  group = group_by_voucher.get_group(index)

  forecast = train_and_forecast(group)

  for_loop_forecast = pd.concat((for_loop_forecast, forecast))

for_loop_forecast
```
